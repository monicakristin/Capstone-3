## A. Business Problem Understanding

Model ini akan membantu pembeli untuk mengukur kewajaran harga yang tercantum, memungkinkan mereka membuat keputusan pembelian yang lebih terinformasi dan sesuai dengan kondisi pasar saat ini.

***Metrics***
Dalam mengevaluasi performa model prediksi harga perumahan di California, digunakan beberapa metrik evaluasi kunci.
* Mean Absolute Error (MAE) digunakan untuk mengukur rataan nilai absolut dari selisih antara prediksi dan nilai aktual.
* MAPE mengukur rataan persentase error dari prediksi terhadap nilai aktual. Metrik ini memberikan pemahaman tentang sejauh mana model dapat memprediksi harga perumahan dengan tepat dalam skala persentase.
* R-squared mengindikasikan seberapa baik variabilitas dalam data dapat dijelaskan oleh model. Semakin mendekati 1, semakin baik model mampu menjelaskan variasi data.
Dengan menggunakan kombinasi metrik evaluasi ini, kita dapat mengukur dan memahami performa model secara holistik, memastikan bahwa model yang dipilih dapat memberikan prediksi harga perumahan yang akurat dan dapat diandalkan.

## B. Data Understanding

Dataset : Harga rumah median untuk distrik-distrik di California yang berasal dari sensus tahun 1990.
Informasi data-datanya: 
1. longitude: Ukuran sejauh mana suatu rumah berada ke barat; nilai yang lebih tinggi berarti lebih ke barat.
2. latitude: Ukuran sejauh mana suatu rumah berada ke utara; nilai yang lebih tinggi berarti lebih ke utara.
3. housingMedianAge: Usia median rumah dalam suatu blok; angka yang lebih rendah menunjukkan bangunan yang lebih baru.
4. totalRooms: Total jumlah ruangan dalam suatu blok.
5. totalBedrooms: Total jumlah kamar tidur dalam suatu blok.
6. population: Total jumlah penduduk tinggal dalam suatu blok.
7. households: Total jumlah rumah tangga, kelompok orang yang tinggal dalam suatu unit rumah, dalam suatu blok.
8. medianIncome: Pendapatan median untuk rumah tangga dalam suatu blok rumah (diukur dalam puluhan ribu Dolar AS).
9. medianHouseValue: Nilai rumah median untuk rumah tangga dalam suatu blok (diukur dalam Dolar AS).
10.oceanProximity: Lokasi rumah terkait dengan samudera/laut.

## G. Modeling

```python
# Import library
from sklearn.model_selection import cross_val_score, KFold,RandomizedSearchCV
import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.metrics import  r2_score, mean_absolute_error, mean_absolute_percentage_error, make_scorer
```

> ### G.1 Model tanpa fitur polinomial

```python
# Mendefinisikan model
linreg = LinearRegression()
knn = KNeighborsRegressor()
ridge = Ridge()
tree = DecisionTreeRegressor(random_state=2020)
rf = RandomForestRegressor(random_state=2020)
xgb = XGBRegressor(random_state=2020)
```

```python
models = [
    ('linear_regression', LinearRegression()),
    ('knn_regressor', KNeighborsRegressor()),
    ('ridge', Ridge()),
    ('decision_tree', DecisionTreeRegressor(random_state=2020)),
    ('random_forest', RandomForestRegressor(random_state=2020)),
    ('xgboost', XGBRegressor(random_state=2020))
]

# List untuk menyimpan hasil evaluasi
scores_mae = []
scores_mape = []
scores_r2 = []

for model_name, model in models:
    crossval = KFold(n_splits=5, shuffle=True, random_state=2020)
    
    # Pipeline untuk scaling pada model linear regression, knn, dan Ridge
    if model_name in ['linear_regression', 'knn_regressor', 'ridge']:
        estimator = make_pipeline(RobustScaler(), model)
    else:
        estimator = model
    
    # Cross-validation untuk masing-masing metrik evaluasi
    score_mae = cross_val_score(estimator, X_train, y_train, cv=crossval, scoring='neg_mean_absolute_error')
    score_mape = cross_val_score(estimator, X_train, y_train, cv=crossval, scoring='neg_mean_absolute_percentage_error')
    score_r2 = cross_val_score(estimator, X_train, y_train, cv=crossval, scoring='r2')
    
    # Mengonversi skor negatif menjadi positif
    score_mae = -score_mae
    score_mape = -score_mape
    
    # Menyimpan hasil evaluasi
    scores_mae.append(score_mae)
    scores_mape.append(score_mape)
    scores_r2.append(score_r2)

    # Menampilkan hasil evaluasi untuk setiap model
    print(f"Model: {model_name}")
    print(f"MAE: Mean = {score_mae.mean()}, Std = {score_mae.std()}")
    print(f"MAPE: Mean = {score_mape.mean()}, Std = {score_mape.std()}")
    print(f"R-squared: Mean = {score_r2.mean()}, Std = {score_r2.std()}")
    print("---------------------------------------")
```

```python
result_df = pd.DataFrame({
    'Model (No Polynomial)': ['Linear Regression', 'KNN Regressor', 'Ridge', 'Decision Tree', 'Random Forest', 'XGBoost'],
    'Mean_MAE': [score.mean() for score in scores_mae],
    'Std_MAE': [score.std() for score in scores_mae],
    'Mean_MAPE': [score.mean() for score in scores_mape],
    'Std_MAPE': [score.std() for score in scores_mape],
    'Mean_R2': [score.mean() for score in scores_r2],
    'Std_R2': [score.std() for score in scores_r2]
})
result_df
```

```python
poly_models_degree2 = [
    ('linear_regression_poly', LinearRegression()),
    ('knn_regressor_poly', KNeighborsRegressor()),
    ('ridge_poly', Ridge()),
    ('decision_tree_poly', DecisionTreeRegressor(random_state=2020)),
    ('random_forest_poly', RandomForestRegressor(random_state=2020)),
    ('xgboost_poly', XGBRegressor(random_state=2020))
]

poly_scores_mae_degree2 = []
poly_scores_mape_degree2 = []
poly_scores_r2_degree2 = []

for model_name, model in poly_models_degree2:
    crossval = KFold(n_splits=5, shuffle=True, random_state=2020)
    
    # Pipeline untuk scaling pada model linear regression, knn, dan Ridge dengan Polynomial Features
    if model_name in ['linear_regression_poly', 'knn_regressor_poly', 'ridge_poly']:
        estimator = make_pipeline(RobustScaler(), PolynomialFeatures(degree=2), model)
    else:
        estimator = make_pipeline(PolynomialFeatures(degree=2), model)
    
    # Cross-validation untuk masing-masing metrik evaluasi
    score_mae = -cross_val_score(estimator, X_train, y_train, cv=crossval, scoring='neg_mean_absolute_error')
    score_mape = -cross_val_score(estimator, X_train, y_train, cv=crossval, scoring='neg_mean_absolute_percentage_error')
    score_r2 = cross_val_score(estimator, X_train, y_train, cv=crossval, scoring='r2')
    
    # Menyimpan hasil evaluasi
    poly_scores_mae_degree2.append(score_mae)
    poly_scores_mape_degree2.append(score_mape)
    poly_scores_r2_degree2.append(score_r2)

    # Menampilkan hasil evaluasi untuk setiap model
    print(f"Model: {model_name}")
    print(f"MAE: Mean = {score_mae.mean()}, Std = {score_mae.std()}")
    print(f"MAPE: Mean = {score_mape.mean()}, Std = {score_mape.std()}")
    print(f"R-squared: Mean = {score_r2.mean()}, Std = {score_r2.std()}")
    print("---------------------------------------")
```

### Algoritma XGBoost

XGBoost (eXtreme Gradient Boosting) adalah algoritma machine learning yang termasuk dalam kategori ensemble learning. Algoritma ini bekerja dengan menggabungkan beberapa model pohon keputusan yang lemah untuk menghasilkan model yang lebih kuat dan akurat

Berikut adalah langkah-langkah dasar dalam algoritma XGBoost:

1. Inisialisasi

Buat model pohon keputusan pertama dengan hanya satu simpul yang mewakili nilai rata-rata dari data target.
2. Iteasi:

Pada setiap iterasi, lakukan langkah-langkah berikut:
Hitung gradien untuk setiap data point, yang menunjukkan arah dan besarnya perubahan yang diperlukan pada prediksi model untuk meningkatkan akurasi.Buat pohon keputusan baru yang fokus pada data point dengan gradien yang tinggi. Tambahkan pohon keputusan baru ke model ensemble. Perbarui bobot setiap pohon keputusan dalam model ensemble.
3. Penghentian:

Hentikan proses iterasi ketika salah satu dari kondisi berikut terpenuhi:
Jumlah iterasi mencapai nilai maksimum yang ditentukan. Akurasi model pada data validasi tidak lagi meningkat.
Berikut adalah beberapa keunggulan XGBoost:

Akurasi tinggi: XGBoost telah terbukti berkinerja sangat baik dalam berbagai kompetisi machine learning dan seringkali mencapai hasil terbaik.
Kecepatan: XGBoost dapat dilatih dengan sangat cepat, bahkan pada dataset yang besar.
Regularisasi bawaan: XGBoost memiliki mekanisme regularisasi built-in yang membantu mencegah overfitting.
Fleksibilitas: XGBoost dapat digunakan untuk berbagai jenis masalah machine learning, termasuk klasifikasi, regresi, dan ranking.
Berikut adalah beberapa kekurangan XGBoost:

Tuning hiperparameter: XGBoost memiliki banyak hiperparameter yang dapat di-tuning untuk meningkatkan performa. Tuning hiperparameter bisa memakan waktu dan membutuhkan keahlian.
Interpretasi: Model XGBoost bisa sulit untuk diinterpretasi, karena terdiri dari banyak pohon keputusan yang kompleks.

## H. Hyperparameter Tuning






Melakukan tahapan hyperparameter tuning untuk model yang dipilih yaitu XGBoost tanpa fitur polinomial

```python
# Membuat pipeline dengan PolynomialFeatures dan XGBRegressor
pipeline = make_pipeline(PolynomialFeatures(degree=3), XGBRegressor(random_state=2020))

param_dist = {
    'xgbregressor__max_depth': list(np.arange(1, 11)),
    'xgbregressor__learning_rate': list(np.arange(1, 100)/100),
    'xgbregressor__n_estimators': list(np.arange(100, 201)),
    'xgbregressor__subsample': list(np.arange(2, 10)/10),
    'xgbregressor__gamma': list(np.arange(1, 11)),
    'xgbregressor__colsample_bytree': list(np.arange(1, 10)/10),
    'xgbregressor__reg_alpha': list(np.logspace(-3, 1, 10))
}

random_search = RandomizedSearchCV(pipeline, param_distributions=param_dist, n_iter=100, 
                                   scoring={'mae': 'neg_mean_absolute_error',
                                            'mape': 'neg_mean_absolute_percentage_error',
                                            'r2': 'r2'},
                                   refit='mae', cv=5, random_state=2020)

random_search.fit(X_train, y_train)
print("Best parameters:", random_search.best_params_)
print("Best MAE score:", -random_search.best_score_)
print("Best MAPE score:", -random_search.cv_results_['mean_test_mape'][random_search.best_index_])
print("Best R-squared score:", random_search.cv_results_['mean_test_r2'][random_search.best_index_])
```

```python
# Mendefinisikan model yang telah dituning untuk model
tuned_model_poly_degree3 = random_search.best_estimator_
y_pred_tuned_poly_degree3 = tuned_model_poly_degree3.predict(X_test)
mae_tuned_poly_degree3 = mean_absolute_error(y_test, y_pred_tuned_poly_degree3)
mape_tuned_poly_degree3 = mean_absolute_percentage_error(y_test, y_pred_tuned_poly_degree3)
r2_tuned_poly_degree3 = r2_score(y_test, y_pred_tuned_poly_degree3)


score_after_tuning_poly_degree3 = pd.DataFrame({
    'MAE': [mae_tuned_poly_degree3],
    'MAPE': [mape_tuned_poly_degree3],
    'R2': [r2_tuned_poly_degree3]
}, index=["XGB Tuning dengan Polynomial Degree=3"])

score_after_tuning_poly_degree3
```
